```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([13,28,31,34,46,50,58,61,72,90])
y = np.array([22,34,43,55,65,79,86,90,94,99])
z = np.array([91,83,75,64,58,43,38,21,10,9])
```

## Line Plot

```python
plt.title("Line Plot")
plt.xlabel("First Value")
plt.ylabel("Second Value")
plt.xticks(range(1,101,5))
plt.yticks(range(1,101,5))
plt.plot(x,y,marker = "o",mfc = "red",color = "black",ls = ":",ms = 7);
plt.grid()
```

```python
plt.title("Line plot")
plt.xlabel("first values")
plt.ylabel("second values")
plt.xticks(range(0,101,5))
plt.yticks(range(0,101,5))
plt.plot(x,z,marker = "H",mfc = "red",color = "black",ls = ":",ms ="7" )
plt.grid()
```

```python
plt.title("Line Plot")
plt.xlabel("First Value")
plt.ylabel("Second Value")
plt.xticks(range(0,101,5))
plt.yticks(range(0,101,5))
plt.plot(x,y,marker = "o",mfc = "red",color = "black",ls = ":",ms = 7,label = "2023");
plt.plot(x,z,marker = "o",mfc = "yellow",color = "black",ls = ":",ms =7,label = "2024" );
plt.legend()
plt.grid()
```

```python
# subplot 
plt.figure(figsize = (16,9))
plt.subplot(2,2,1)

plt.title("Line plot")
plt.xlabel("first values")
plt.ylabel("second values")
plt.xticks(range(0,101,5))
plt.yticks(range(0,101,5))
plt.plot(x,z,marker = "o",mfc = "red",color = "black",ls = ":",ms ="7" )
plt.grid()

plt.subplot(2,2,2)
plt.title("Line Plot")
plt.xlabel("First Value")
plt.ylabel("Second Value")
plt.xticks(range(1,101,5))
plt.yticks(range(1,101,5))
plt.plot(x,y,marker = "o",mfc = "red",color = "black",ls = ":",ms = 7);
plt.grid()
```

## Bar Plot

```python
courses = ["Java","C","Python","R","Kotlin"]
students = [50,66,87,12,24]
plt.title("2024")
plt.xlabel("Students")
plt.ylabel("Courses")
plt.yticks(range(0,100,5))
plt.bar(courses,students,width = 0.300,color = "red",edgecolor = "black",linewidth = 2.5);
```

## Pie Chart

```python
color = ["orange","cyan","green","yellow","purple"]
plt.pie(students , labels = courses,autopct = "%.f%%",labeldistance = 1.2,startangle = 90,shadow = True,explode = [0.14,0,0,0,0]
       ,colors = color);
```

# Box Plot 

```python
marks = [45,67,69,74,53,65,56,67,99] 
plt.boxplot(marks);
q3 = np.percentile(marks,75)
q1 = np.percentile(marks,25)
iqr = q3 - q1
```

```python
ub = q3 + (1.5 * iqr)
lb = q1 - (1.5 * iqr)
ub , lb
```
